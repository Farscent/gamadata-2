```python
import math

# create a function that can visualize the trend of any feature from the dataset in a single canvas
def plot_feature_trend_by_province_in_canvas(df, feature, provinces, title, n_cols=6):
    """
    Plots the trend of a specific feature for each province in a single rectangular canvas.

    Parameters:
    - df: pandas DataFrame containing the data.
    - feature: Column name of the feature to plot.
    - provinces: List or array of unique province names.
    - title: Title for the entire plot.
    - n_cols: Number of columns in the subplot grid (default is 6).
    """
    n_provinces = len(provinces)
    n_rows = math.ceil(n_provinces / n_cols)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 4, n_rows * 3), sharex=True, sharey=True)
    axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

    # Set overall title for the figure
    fig.suptitle(title, fontsize=16, y=1.02)

    for i, province in enumerate(provinces):
        province_data = df[df['Provinsi'] == province]
        axes[i].plot(province_data['Tahun'], province_data[feature], marker='o', label=province, color='blue')
        axes[i].set_title(province, fontsize=10)
        axes[i].tick_params(axis='x', rotation=45, labelsize=8)
        axes[i].tick_params(axis='y', labelsize=8)
        axes[i].grid(True, linestyle='--', alpha=0.7)

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout(rect=[0, 0.03, 1, 0.98])  # Adjust layout to make space for suptitle
    plt.show()

# Example usage of the updated function
feature_to_plot = 'Indeks Keparahan Perdesaan Semester 2'
plot_feature_trend_by_province_in_canvas(
    df=df_merged,
    feature=feature_to_plot,
    provinces=provinces,
    title='Trend of Indeks Keparahan Perdesaan Semester 2 by Province (2015-2024)'
)
```

```python
import pandas as pd
import numpy as np

def drop_column(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
    """
    Drops a specified column from the DataFrame if it exists.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column to drop.

    Returns:
        pd.DataFrame: The DataFrame with the specified column dropped.
    """
    if column_name in df.columns:
        df = df.drop(columns=[column_name])
        print(f"Column '{column_name}' dropped successfully.")
    else:
        print(f"Column '{column_name}' not found in DataFrame, skipping drop.")
    return df

def impute_year_midpoint(df: pd.DataFrame, target_year: int, feature_columns: list) -> pd.DataFrame:
    """
    Imputes missing values for a specific year in specified feature columns
    by taking the midpoint between the previous and next year's values for each province.

    Args:
        df (pd.DataFrame): The input DataFrame.
        target_year (int): The year for which to impute missing values (e.g., 2023).
        feature_columns (list): A list of column names to impute.

    Returns:
        pd.DataFrame: The DataFrame with imputed values for the target year.
    """
    print(f"\n--- Imputing missing values for year {target_year} with midpoint ---")
    provinces = df['Provinsi'].unique()

    for col in feature_columns:
        if col not in df.columns:
            print(f"Warning: Feature column '{col}' not found. Skipping imputation for it.")
            continue

        # Temporarily set the target_year's values to NaN if they are not already
        # This ensures consistent re-imputation if the function is run multiple times
        df.loc[df['Tahun'] == target_year, col] = np.nan

        for province in provinces:
            value_prev_year = df.loc[(df['Provinsi'] == province) & (df['Tahun'] == target_year - 1), col].values
            value_next_year = df.loc[(df['Provinsi'] == province) & (df['Tahun'] == target_year + 1), col].values

            # Check if both previous and next year's values are available and not NaN
            if len(value_prev_year) > 0 and not pd.isna(value_prev_year[0]) and \
               len(value_next_year) > 0 and not pd.isna(value_next_year[0]):
                midpoint = (value_prev_year[0] + value_next_year[0]) / 2
                df.loc[(df['Provinsi'] == province) & (df['Tahun'] == target_year), col] = midpoint
                # print(f"  Imputed '{col}' for {province} in {target_year} with {midpoint:.2f}")
            else:
                print(f"  Warning: Could not impute '{col}' for {province} in {target_year} due to missing data in {target_year - 1} or {target_year + 1}.")
    print(f"--- Finished year {target_year} imputation ---")
    return df

def impute_province_zero(df: pd.DataFrame, target_province: str, feature_columns: list) -> pd.DataFrame:
    """
    Imputes missing values for a specific province in specified feature columns with zero.
    This is typically used for features that are not applicable to a given province (e.g., rural data in urban areas).

    Args:
        df (pd.DataFrame): The input DataFrame.
        target_province (str): The name of the province to impute (e.g., 'DKI JAKARTA').
        feature_columns (list): A list of column names to impute with zero.

    Returns:
        pd.DataFrame: The DataFrame with specified values imputed to zero.
    """
    print(f"\n--- Imputing missing values for {target_province} with 0 ---")
    if target_province not in df['Provinsi'].unique():
        print(f"Warning: Province '{target_province}' not found in DataFrame. Skipping imputation.")
        return df

    for col in feature_columns:
        if col not in df.columns:
            print(f"Warning: Feature column '{col}' not found. Skipping imputation for it.")
            continue
        
        # Only impute if there are actual NaNs for this province in this column
        if df.loc[(df['Provinsi'] == target_province), col].isnull().any():
            df.loc[(df['Provinsi'] == target_province), col] = 0
            print(f"  Imputed '{col}' for {target_province} with 0.")
        else:
            print(f"  No missing values found in '{col}' for {target_province}, skipping.")
    print(f"--- Finished {target_province} imputation ---")
    return df

# --- Initial Data Loading and Preprocessing (to ensure df_filtered is ready) ---
# Load the CSV file into a DataFrame
df_raw = pd.read_csv('Full Data From 2015.csv')

# Filter for years 2015 to 2024
df_filtered = df_raw[(df_raw['Tahun'] >= 2015) & (df_raw['Tahun'] <= 2024)].copy()

# Drop the 'Unnamed: 0' column
df_filtered = df_filtered.drop(columns=['Unnamed: 0'])

# Reset the index
df_filtered = df_filtered.reset_index(drop=True)

print("Initial DataFrame head:")
print(df_filtered.head())
print("\nInitial missing values count:")
print(df_filtered.isnull().sum()[df_filtered.isnull().sum() > 0])

# --- Demonstration of using the functions ---

# Example 1: Drop 'UMP' column
df_filtered = drop_column(df_filtered, 'UMP')

# Example 2: Impute 'Garis Kemiskinan Makanan Total Semester 2' for 2023 with midpoint
# (Assuming this is the 'blue circle' type missingness)
df_filtered = impute_year_midpoint(df_filtered, 2023, ['Garis Kemiskinan Makanan Total Semester 2'])

# Example 3: Impute DKI Jakarta's rural-related columns with 0
# (Assuming these are the 'red circle' type missingness, which also covers 'green circle' if applicable)
dki_jakarta_rural_cols = [
    'Garis Kemiskinan Non Makanan Perdesaan Semester 1',
    'Garis Kemiskinan Non Makanan Perdesaan Semester 2',
    'Indeks Keparahan Perdesaan Semester 1',
    'Indeks Keparahan Perdesaan Semester 2'
]
df_filtered = impute_province_zero(df_filtered, 'DKI JAKARTA', dki_jakarta_rural_cols)

# Final check for any remaining missing values after all operations
print("\nFinal missing values count after all operations:")
print(df_filtered.isnull().sum()[df_filtered.isnull().sum() > 0])
print(f"Total remaining missing values: {df_filtered.isnull().sum().sum()}")
```
