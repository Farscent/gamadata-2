```python
provinsi_df1 = set(garis_kemiskinan_makanan['Provinsi'].unique())
provinsi_df2 = set(garis_kemiskinan_non_makanan['Provinsi'].unique())
provinsi_df3 = set(gini_ratio['Provinsi'].unique())
provinsi_df4 = set(ump_long['Provinsi'].unique())
provinsi_df5 = set(inflation['Provinsi'].unique())
provinsi_df6 = set(pdrb['Provinsi'].unique())
provinsi_df7 = set(indeks_keparahan['Provinsi'].unique())
```

```python
provinsi_df1 == provinsi_df2 == provinsi_df3 == provinsi_df4 == provinsi_df6 == provinsi_df7
```

```python
provinsi_sets = [
    provinsi_df1, provinsi_df2, provinsi_df3,
    provinsi_df4, provinsi_df6, provinsi_df7
]

names = [
    "provinsi_df1", "provinsi_df2", "provinsi_df3",
    "provinsi_df4", "provinsi_df5", "provinsi_df6", "provinsi_df7"
]

reference = provinsi_sets[0]

for i in range(1, len(provinsi_sets)):
    if provinsi_sets[i] != reference:
        print(f"{names[i]} berbeda dari provinsi_df1")

        diff_1_to_X = reference - provinsi_sets[i]
        if diff_1_to_X:
            print(f"  - Provinsi yang ada di provinsi_df1 tapi tidak di {names[i]}:")
            print(f"    {diff_1_to_X}")

        diff_X_to_1 = provinsi_sets[i] - reference
        if diff_X_to_1:
            print(f"  - Provinsi yang ada di {names[i]} tapi tidak di provinsi_df1:")
            print(f"    {diff_X_to_1}")
```
